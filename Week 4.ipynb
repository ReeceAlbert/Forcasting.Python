```python
# test horizon, how long my test data will be
test_horizon = ForecastingHorizon(y_test.index,is_relative=False)
test_horizon
```

```python
# Forcasting 
model_historic_mean = NaiveForecaster(strategy='mean')
fit_historic_mean = model_historic_mean.fit(y_train)
```

```python
fcast_historic_mean = fit_historic_mean.predict(test_horizon)
fcast_historic_mean
```

```python
y_train.plot(label='Training')
y_test.plot(label='Testing')
fcast_historic_mean.plot(label='Historic Mean Forecast')
plt.legend()
plt.show()
```

```python
# historical mean from a specific window
model12_momean = NaiveForecaster(strategy='mean',window_length=12) # 12 periods 
fit_12mo = model12_momean.fit(y_train)
```

```python
def drift_forecast(timeSeries,num_of_forecast_periods,frequency): 
    import pandas as pd
    from dateutil.relativedelta import relativedelta    
    try:
        T = len(timeSeries)
        df = timeSeries
        df.index=df.index.astype('datetime64[ns]')
        h = num_of_forecast_periods     
        first_value = df.values[0][0]     # here [1] finds the first row, [0] finds the first element of the first array
        last_value = df.values[-1][0]     # here [-1] finds the last row, [0] finds the first element of the last array
        first_dt = df.index[0]
        last_dt= df.index[-1]
        drift_horizon = len(df) - 1
        drift = (last_value - first_value) / drift_horizon            
        if frequency == 12:
            forecast_horizon = pd.date_range(start=last_dt+ relativedelta(months=1), end=last_dt + relativedelta(months=h), freq='MS')
        elif frequency == 4:
            forecast_horizon = pd.date_range(start=last_dt+ relativedelta(months=1), end=last_dt + relativedelta(months=h), freq='QS')
        elif frequency == 1:
            forecast_horizon = pd.date_range(start=last_dt+ relativedelta(months=1), end=last_dt + relativedelta(months=h), freq='YS')
        else:
            error_message = 'Frequency input must be numerical 1 (i.e., yearly), 4 (i.e., quarterly) or 12 (i.e., monthly)! Check Input Time Series Index Freq'
        fcast = pd.DataFrame(forecast_horizon, columns=[df.index.name])
        fcast.set_index(df.index.name, inplace=True)
        fcast['Forecast'] = [last_value + i * (last_value - first_value) * 1/(T-1) for i in range(len(fcast))]
        return fcast
    except:
        print(error_message, 'Check Inputs!')
```

```python
def accuracy(y_test,y_pred):
    from statsmodels.tools.eval_measures import rmse, rmspe, meanabs, mse
    from sktime.performance_metrics.forecasting import mean_absolute_percentage_error  
    RMSE = rmse(y_test,y_pred)
    RMSPE = rmspe(y_test,y_pred)
    MAE = meanabs(y_test,y_pred)
    MSE =  mse(y_test,y_pred)
    MAPE = mean_absolute_percentage_error(y_test,y_pred)    
    print(' Error Statistics : RMSE %f' % RMSE, '    RMSPE %f' % RMSPE, '    MAE %f' % MAE, '    MSE %f' % MSE , '      MAPE %f' % MAPE)
```
