```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MSACSR.csv',index_col=0,parse_dates=True)
df.head()
```

```python
rolling_3mo = df['MSACSR'].rolling(window=3,center=True)
rolling_6mo = df['MSACSR'].rolling(window=6,center=True)
rolling_12mo = df['MSACSR'].rolling(window=12,center=True)
```

```python
mov_avg_3 = rolling_3mo.sum()
mov_avg_6 = rolling_6mo.sum()
mov_avg_12 = rolling_12mo.sum()
```

```python
df.plot(label='Original Data')
mov_avg_3.plot(label='Moving Avg of 3')
mov_avg_6.plot(label='Moving Avg of 6')
mov_avg_12.plot(label='Moving Avg of 12')
plt.legend()
plt.show()
```
